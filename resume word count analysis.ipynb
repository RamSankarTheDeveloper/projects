```python
text=text.lower()
text = re.sub('[^A-Za-z0-9 ]+', ' ', text)
```

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize# Passing the string text into word tokenize for breaking the sentences
token = word_tokenize(text)
#token
```

```python
from nltk.corpus import stopwords
a = set(stopwords.words('english'))
stopwords = [x for x in token if x not in a]
#print(stopwords)
```

```python
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()
#[lst.stem(x) for x in stopwords]
```

```python
# quick test code
#unique_list = [list(x) for x in stopwords.copy()]
stopwords2 = stopwords.copy()
for index1, item1 in enumerate(stopwords2):
    for index2, item2 in enumerate(stopwords2):
        if index1 == index2:
            continue
        else:
            if (item1[:4] in item2[:4]):
                
                if len(item1)>len(item2):
                    stopwords2[index2] = item1
                else:
                    stopwords2[index1] = item2
#stopwords2
```

```python
stopwords2 = stopwords.copy()
count_list = [[x,stopwords2.count(x)] for x in stopwords2 ]
unique_count_list=[]
for x in count_list:
    if x not in unique_count_list:
        unique_count_list.append(x)
#unique_count_list
```

```python
string = [x[0] for x in unique_count_list]
count = [x[1] for x in unique_count_list]
string2 = []
count2 = []
for i,x in enumerate(count):
    for i,x in enumerate(count):
        if x == max(count):
            string2.append(string[i])
            count2.append(count[i])
            del string[i]
            del count[i]
count2
```

```python
for i ,x in enumerate(string2):
        if x[0] in "0123456789":
            del string2[i]
            del count2[i]
```

```python
from sklearn.cluster import KMeans
wcsslist=[]

for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init = 'k-means++',random_state=42)
    kmeans.fit(x)
    wcsslist.append(kmeans.inertia_)
plt.plot(range(1,11),wcsslist)
plt.title('the elbow method graph')
plt.xlabel('number of clusters(k)')
plt.ylabel('wcss_list')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer

#tokenizer to remove unwanted elements from out data like symbols and numbers
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(lowercase=True,stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)
text_counts= cv.fit_transform(stringseries)
print(text_counts)
```

```python
kmeans=KMeans(n_clusters=2, init='k-means++',random_state=42)
y_predict=kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_predict==0,0], x[y_predict==0,1],s=100,c='blue',label='cluster 1')#for first cluster
plt.scatter(x[y_predict==1,0], x[y_predict==1,1],s=100,c='green',label='cluster 2')#for second cluster
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='centroid')
plt.title('clusters')
plt.xlabel('words(201)')
plt.ylabel('count')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(text_counts,x[1], test_size=0.05, random_state=1)
```

```python
from sklearn.naive_bayes import MultinomialNB
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Generation Using Multinomial Naive Bayes
clf = MultinomialNB().fit(X_train, y_train)
predicted= clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, predicted))
```

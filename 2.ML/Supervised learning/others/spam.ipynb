```python
import pandas as pd
df=pd.read_table('SMSSpamCollection',header=None, names=['class','sms'])
df.head()
```

```python
df['class'].value_counts()
```

```python
df['label']=df['class'].map({'ham':0,'spam':1})
df
```

```python
x=df['sms']
y=df.label
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
```

```python
vect.fit(xtrain)
vect.vocabulary_
```

```python
vect = CountVectorizer(stop_words="english")
```

```python
vectorizer = vect.fit(x)
vect.vocabulary_
```

```python
xtrain_transformed = vect.transform(xtrain)
xtest_transformed = vect.transform(xtest)
```

```python
print(xtrain_transformed)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix

mnb=MultinomialNB()

naive_model = mnb.fit(xtrain_transformed,ytrain)
y_pred_class= mnb.predict(xtest_transformed)
print(classification_report(ytest,y_pred_class))
```

```python
import joblib
fin_model = naive_model
joblib.dump(fin_model, "naive_model.joblib")
joblib.dump(vectorizer, "countvectorizer.joblib")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('headbrain.csv')
data
```

```python
x=data['Head Size(cm^3)'].values
y=data['Brain Weight(grams)'].values
#makes a numpy array with those values
```

```python
x
```

```python
y
```

```python
n=len(x)
x=x.reshape(n,1)
x
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,train_size=0.8)#test size too.
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
type(model)
```

```python
model.fit(xtrain,ytrain)#xtrain, ytrain inputted, regression line created, lowest loss fuction figured out, changed regression line tested the data and calculated attributes
```

```python
y_predicted=model.predict(xtest)
y_predicted
```

```python
r2score=model.score(x,y)
r2score
```

```python
#m=slope=model coefficient
m=model.coef_
m
```

```python
c=model.intercept_
c
```

```python
model.predict([[123]])
```

```python
model.predict(xtest)
```

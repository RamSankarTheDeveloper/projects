1.what is seaborn ?

seaborn is a python library that plots/visualize datasets

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#2how can we seee all dataset available in seaborn library?
sns.get_dataset_names()
```

```python
#3.What is the Seaborn function for colouring plots
df=sns.load_dataset('penguins')
sns.relplot(data=df,x='bill_length_mm',y='bill_depth_mm',hue='species')
plt.show()
```

```python
#4.How do you plot a histogram in Seaborn?
import matplotlib.pyplot as plt
sns.histplot('penguins',x=df['body_mass_g'])
plt.show()
```

```python
#5.How to change the minimum number of ticks of MaxNLocator in Seaborn?


sns.histplot('penguins',x=df['body_mass_g'])
sns.yaxis.set_major_locator(plt.MaxNLocator(3, min_n_ticks=3))
plt.MaxNLocator(2)
plt.show()
```

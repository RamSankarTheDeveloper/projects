3.Read all product sales data and show itusing a multiline plotDisplay the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).The graph should look like thi

```python
plt.plot(df['month_number'],df['facecream'],marker='o',linestyle='dashed',label='face cream',color='red',linewidth="1")
plt.plot(df['month_number'],df['facewash'],marker='o',linestyle='dashed',label='facewash',color='blue',linewidth="1")
plt.plot(df['month_number'],df['toothpaste'],marker='o',label='toothpaste',linestyle='dashed',color='green',linewidth="1")
plt.plot(df['month_number'],df['bathingsoap'],label='bathing soap',marker='o',linestyle='dashed',color='yellow',linewidth="1")
plt.plot(df['month_number'],df['shampoo'],label='shampoo',marker='o',linestyle='dashed',color='magenta',linewidth="1")
plt.plot(df['month_number'],df['moisturizer'],label='moisturizer',marker='o',linestyle='dashed',color='cyan',linewidth="1")
plt.xlabel('month number')
plt.ylabel('sold units number')
plt.legend()
plt.show()
```

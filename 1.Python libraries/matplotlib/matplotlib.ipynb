```python
import math
import numpy as np
```

```python
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)
#plt.show
plt.plot(x,z)
```

```python
#labelling
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title('X-Y plot')
plt.plot(x,y)
```

```python
plt.plot(x,y,'r*')
plt.plot(x,z,'go')
```

```python
#legends
plt.plot(x,y,label='sin wave',linewidth=5, marker='x')
plt.plot(x,z,label='cos wave',linewidth=5, marker='o')
plt.legend()
```

```python
days=[3,4,5,6,7,8]
sleeping=[7,8,6,11,7,5]
eating=[2,3,4,3,2,4]
working=[7,8,7,2,2,2]
playing=[8,5,7,8,13,6]
plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='r',label='working',linewidth=5)
plt.plot([],[],color='k',label='playing',linewidth=5)
plt.stackplot(days,sleeping, eating,working,playing,colors=['m','c','r','k'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('stack plot')
plt.legend()
```
